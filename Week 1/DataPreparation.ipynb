```python
import pandas as pd
```

```python
df = pd.read_csv("PD 2023 Wk 1 Input.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Split the Transaction Code to extract the letters at the start of the transaction code. 
# These identify the bank who processes the transaction and Rename the new field with the Bank code 'Bank'
df['Bank'] = df['Transaction Code'].apply(lambda x: x.split('-',1)[0])
```

```python
df.Bank
```

```python
# Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values.
df['Online or In-Person'] = df['Online or In-Person'].apply(lambda x: 'Online' if x==1 else 'In-Person')
```

```python
df['Online or In-Person'].head()
```

```python
from datetime import datetime
```

```python
# Change the date to be the day of the week
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df['Transaction Date'] = df['Transaction Date'].apply(lambda x: x.strftime('%A'))
```

```python
df['Transaction Date']
```

Different levels of detail are required in the outputs. You will need to sum up the values of the transactions in three ways (help):

    1. Total Values of Transactions by each bank
    2. Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
    3. Total Values by Bank and Customer Code

```python
# 1.
output_1 = df.groupby(by=['Bank']).sum()
output_1 = output_1.drop('Customer Code', axis=1)
```

```python
# 2.
output_2 = df.groupby(by=['Bank', 'Transaction Date','Online or In-Person']).sum()
output_2 = output_2.drop('Customer Code', axis=1)
```

```python
# .3
output_3 = df.drop(['Transaction Date','Online or In-Person'], axis=1)
output_3 = output_3.groupby(by=['Bank', 'Customer Code']).sum()
```
